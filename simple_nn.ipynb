```python
# building a basic nn using pytorch to solve a delivery problem

"""Prepare delivery data, the distances and times from past orders.
Build a simple neural network using PyTorch (just one neuron!).
Train it to find the relationship between distance and delivery time.
Predict whether you can make that 7-mile delivery in time."""
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

import helper_utils

# This line ensures that results are reproducible and consistent every time.
torch.manual_seed(42)
```

```python
# Distances in miles for recent bike deliveries
distances = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)

# Corresponding delivery times in minutes
times = torch.tensor([[6.96], [12.11], [16.77], [22.21]], dtype=torch.float32)
```

```python
# a model with a single linear layer (one neuron)
model = nn.Sequential(nn.Linear(1,1)) # 1 input and 1 output
# 1 weight and 1 bias in the linear layer

# define loss function and optimizer
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
# model.paramters() gets the weights and biases of the model
```

```python
for epoch in range(500):

    optimizer.zero_grad()  # resets the optimiser, that is zeros the gradients
    outputs = model(distances)  # forward pass: compute predicted y (time) by passing x (distances) to the model
    loss = loss_function(outputs, times)  # compute loss
    # times is the actual delivery times
    loss.backward()  # calcualte the adustments required in weights
    # basically apply chain rule goigng backwards through the network
    optimizer.step()  # update weights
    # printing loss after every 50 epochs
    if (epoch+1) % 50 == 0:
        print(f'Epoch [{epoch+1}/500], Loss: {loss.item():.4f}')
```
